## Pandas - Parte I
---

```python
#Carregue os dados da pnad, criando um dataframe pnad e mostre as suas 5 primeiras linhas.
pnad_df = pd.read_csv("pnad.csv")
pnad_df.head()
```

```python
#Mostre tipo de dado de cada coluna do dataframe criado.
pnad_df.dtypes
```

```python
#Mostre o tamanho do dataframe (quantas observações e quantas variáveis).
pnad_df.shape
```

```python
#Mostre as últimas 5 observações do dataframe criado.
pnad_df.tail()
```

```python
#Mostre apenas a variável idade.
#Primeira maneira e a recomendada.
print(pnad_df['idade'])
#Segunda maneira e não recomendada devido espaços e acentos.
print(pnad_df.idade)
```

---
# Medidas de Tendência Central - Parte II
---

```python
#Mostre a idade média.
pnad_df['idade'].mean()
```

```python
#Mostre o resumo estatístico da idade.
pnad_df['idade'].describe()
```

```python
#Mostre o rendimento médio.
pnad_df['rendimento'].mean()
```

```python
#Mostre o rendimento médio das pessoas com maioridade.
#Dica: use o comando query().
pnad_df.query("rendimento >= 18")['rendimento'].mean()
```

```python
#Calcule a media geometrica e harmonica das pessoas com maioridade.
#Dica: use o comando query().

#importanto o modulo do scipy para calculo da media geometrica e harmonica
from scipy import stats

geometric_mean = stats.gmean(pnad_df.query("rendimento >= 18")['rendimento'],axis=0)
print("Media Geometrica do rendimento:", geometric_mean)

#calculo da media harmonica
harmonic_mean = stats.hmean(pnad_df.query("rendimento >= 18")['rendimento'],axis=0)
print("Media Harmonica do rendimento:", harmonic_mean)
```

```python
#Calcule a média de rendimentos para cada um dos sexos.
#Dica: use o comando groupby().

pnad_df.groupby('sexo')['rendimento'].mean()
```

---
# Medidas de Dispersão ou Variabilidade - Parte III
---

```python
#Qual é o desvio padrão do rendimento das pessoas com maioridade.

pnad_df.query("rendimento >= 18")['rendimento'].std()
```

```python
#Calcule a variância dos rendimentos para cada um dos sexos.

pnad_df.groupby('sexo')['rendimento'].var()
```

```python
#Calcule a amplitudade do rendimento das pessoas com maioridade.
amplitude = pnad_df.query("rendimento >= 18")['rendimento'].max() - pnad_df.query("rendimento >= 18")['rendimento'].min()
print(amplitude)
```

```python
#Calcule o desvio padrão dos rendimentos para cada uma das cores/raças em cada um dos sexos.
#Dica: use o comando groupby().

pnad_df.groupby('cor')['rendimento'].std()
```

# Visualização gráfica de dados - Parte IV
---

```python
#Defina o estilo dos gráficos "darkgrid" do seaborn.
sns.set(style='darkgrid')
```

```python
#Crie um histograma apenas dos rendimentos menores do que R$10.000.
#Dica: use o comando query().

sns.distplot(pnad_df.query('rendimento < 10000')['rendimento'])
```

```python
#Crie um histograma das idades
sns.distplot(pnad_df['idade'])
```

```python
#Crie um grafico de setores para a cores/raças.
#Dica: use o comando value_counts().

cores_racas = pnad_df['cor'].value_counts()
valores = cores_racas.index

plt.pie(cores_racas, labels=valores, autopct='%1.1f%%', 
        shadow=True, startangle=90, colors=sns.color_palette('muted'))
```

```python
##Crie um grafico de setores para o sexo.
#Dica: use o comando value_counts().

sexo = pnad_df['sexo'].value_counts()
valores = sexo.index

plt.pie(sexo, labels=valores, autopct='%1.1f%%', 
        shadow=True, startangle=90, colors=sns.color_palette('muted'))
```

```python
#Crie um histograma dos rendimentos entre 1000 e 5000 reais.

sns.distplot(pnad_df.query('rendimento > 1000 & rendimento < 5000')['rendimento'])
```

```python
#Crie um gráfico de barras para 10 primeiras idades.
sns.catplot(data=pnad_df.head(10), kind='count', x = 'idade')
```

```python
#Faça um diagrama de dispersão entre idade e rendimento.
sns.catplot(x='idade', y='rendimento', data=pnad_df, kind='swarm')
```

```python
#Faça um boxplot da idade.
sns.boxplot(data=pnad_df['idade'])
```

```python
#Faça boxplots dos rendimentos para cada um dos estados (variável "uf").
plt.figure(figsize=(15, 10))
sns.boxplot('rendimento', 'uf', data=pnad_df, palette='Accent', orient='h')
```
